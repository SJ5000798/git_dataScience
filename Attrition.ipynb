```python
df.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1,inplace=True)
```

```python
df.dtypes
```

## Label encoding for Catagorical variables

```python
df['Attrition_1'] = np.nan
df.loc[df['Attrition'] == "Yes",'Attrition_1'] = 1
df.loc[df['Attrition'] == "No",'Attrition_1'] = 0
```

```python
df.drop('Attrition',axis=1,inplace=True)
```

```python
colname=['BusinessTravel', 'Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']
##for processing the data
from sklearn import preprocessing
#creating dictionary"""
le={}
for x in colname:
    le[x]=preprocessing.LabelEncoder()  ## generating levels
    
for x in colname:     ### assigning levels
    df[x]=le[x].fit_transform(df[x])
    
df.head()
```

```python
X=df.copy()
```

```python
Y=X.pop('Attrition_1')
```

```python
X.shape
```

```python
Y.shape
```

```python
Y=Y.astype(int)
```

### Creating testing data set for cross validation

```python
#Creating training and testing set
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.3,
                                                 random_state = 10)
```

###  Model Building:-Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
#creating model
classifier=LogisticRegression()
##fitting training data to the model
classifier.fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
y_train_pred = classifier.predict(X_train)
print(confusion_matrix(Y_train,y_train_pred))
print(classification_report(Y_train,y_train_pred))
acc=accuracy_score(Y_train,y_train_pred)
print("Accuracy of model: ",acc)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
y_pred = classifier.predict(X_test)
print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
acc=accuracy_score(Y_test,y_pred)
print("Accuracy of model: ",acc)
```
